```
#Mesh plot of Guassian Kernel
from matplotlib import cm

x=np.linspace(-10,10, num=100)
y=np.linspace(-10,10, num=100)
x, y = np.meshgrid(x, y)
sigma=1
z=np.exp(-(np.square(x)+np.square(y))/(2*(np.pi)*np.square(sigma))) #Gaussian Equation
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=1")

sigma=2
z=np.exp(-(np.square(x)+np.square(y))/(2*(np.pi)*np.square(sigma))) #Gaussian Equation
ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=2")

sigma=3
z=np.exp(-(np.square(x)+np.square(y))/(2*(np.pi)*np.square(sigma))) #Gaussian Equation
ax = fig.add_subplot(133, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=3")
```

```
#Mesh plot of DoG Kernel. Derivative is computed in both x and y direction so both plots can be seen as an output
sigma=1
z = -(x/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(331, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=1 DoG with respect to X")
z = -(y/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
ax = fig.add_subplot(334, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=1 DoG with respect to Y")

sigma=2
z = -(x/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
ax = fig.add_subplot(332, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=2 DoG with respect to X")
z = -(y/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
ax = fig.add_subplot(335, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=2 DoG with respect to Y")

sigma=3
z = -(x/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
ax = fig.add_subplot(333, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=3 DoG with respect to X")
z = -(y/(2*(np.pi)*((np.square(sigma))*(np.square(sigma)))))*(np.exp(-(np.square(x)+np.square(y))/(2*(np.square(sigma)))))
ax = fig.add_subplot(336, projection='3d')
ax.plot_surface(x,y,z,cmap=cm.jet)
ax.set_title("sigma=3 DoG with respect to Y")
```
