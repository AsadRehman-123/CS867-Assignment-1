```
#A canny edge detector is used to find edges of two images
image1 = cv2.imread("07.jpg",0)
image2 = cv2.imread("08.jpg",0)
edges1 = cv2.Canny(image1, 100, 200)
edges2 = cv2.Canny(image2, 100, 200)
plt.rcParams['figure.figsize'] = (20.0, 16.0) # set default size of plots
fig = plt.figure()
fig.add_subplot(2, 2, 1)
plt.imshow(image1,cmap=plt.cm.gray)
plt.title("GrayScale")
fig.add_subplot(2, 2, 2).imshow(edges1,cmap=plt.cm.gray)
plt.title("Canny edge detector")
fig.add_subplot(2, 2, 3)
plt.imshow(image2,cmap=plt.cm.gray)
plt.title("GrayScale")
fig.add_subplot(2, 2, 4).imshow(edges2,cmap=plt.cm.gray)
plt.title("Canny edge detector")
```
